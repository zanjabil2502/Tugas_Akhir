```python
import pandas as pd
import os
import librosa as lb
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import soundfile as sf
import random
import shutil
```

```python
if os.path.exists('noise_segmentation')==False:
    os.mkdir('noise_segmentation')
```

```python
def slice_data(start, end, raw_data,  sample_rate):
    max_ind = len(raw_data) 
    start_ind = min(int(start * sample_rate), max_ind)
    end_ind = min(int(end * sample_rate), max_ind)
    return raw_data[start_ind: end_ind]
```

```python
for file in os.listdir('sample_noise'):
    filename = os.path.join('sample_noise',file)
    wav,sr = lb.load(filename,sr=None)
    wav = lb.util.normalize(wav)
    wav = lb.resample(wav, orig_sr=sr, target_sr=16000)
    duration = lb.get_duration(wav,sr)
    duration = round(duration)
    for num in range(0,duration):
        start = num*0.3
        end = (num+1)*0.3
        data = slice_data(start,end,raw_data=wav,sample_rate=sr)
        fn = file.split('.')[0]+'_'+str(num)+'.wav'
        path = os.path.join('noise_segmentation',fn)
        print(fn)
        sf.write(file=path, data=data, samplerate=16000)
```
