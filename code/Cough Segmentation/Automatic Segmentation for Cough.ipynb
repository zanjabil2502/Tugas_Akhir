```python
import pandas as pd
import os
import librosa as lb
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import soundfile as sf
```

```python
dataset = 'wav_normalized'
Scaler = MinMaxScaler(feature_range=(0,1))
```

```python
df_compare = pd.read_csv('Compare_dataset.csv')
df_compare.head()
```

```python
folder = ['train','devel','test']
type_folder = ['positive','negative']

if os.path.exists('Segment')==False:
    os.mkdir('Segment')
    for i in folder:
        os.mkdir(os.path.join('Segment',i))
        for j in type_folder:
            os.mkdir(os.path.join('Segment',i,j))
```

```python
def slice_data(start, end, raw_data,  sample_rate):
    max_ind = len(raw_data) 
    start_ind = min(int(start * sample_rate), max_ind)
    end_ind = min(int(end * sample_rate), max_ind)
    return raw_data[start_ind: end_ind]
```

```python
for index,row in df_compare.iterrows():
    file = row['filename']
    label = row['label']
    filename = os.path.join(dataset,file)
    wav, sr = lb.load(filename, sr=None)

    rms = lb.feature.rms(y=wav)
    times = lb.times_like(rms, sr=16000)
    rms = rms.reshape(rms.shape[1],rms.shape[0])
    rms_norm = Scaler.fit_transform(rms)

    index = 0
    time_cut = []
    start = []
    end = []
    for value in rms_norm:
        if value > 0.09:
            time_cut.append(times[index-3])
            try:
                if rms_norm[index+1] < 0.09:
                    time_cut.append(times[index+3])

                    start.append(time_cut[0])
                    end_index = len(time_cut)
                    end.append(time_cut[end_index-1])
                    time_cut = []
            except:
                time_cut.append(times[index])

                start.append(time_cut[0])
                end_index = len(time_cut)
                end.append(time_cut[end_index-1])
                time_cut = []

        index=index+1

    for num in range(len(start)):
        sliced_data = slice_data(start=start[num], end=end[num], raw_data=wav, sample_rate=16000)
        fdr = file.split('_')[0]
        fn = file.split('.')[0]+'-'+str(num)+'.wav'
        path = os.path.join('Segment',fdr,label,fn)
        print(fdr,label,fn)
        sf.write(file=path, data=sliced_data, samplerate=16000)
```
