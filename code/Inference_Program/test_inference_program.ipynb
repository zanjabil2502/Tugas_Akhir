```python
import os
import pandas as pd
```

```python
df_data = pd.read_csv('test.csv')
df_data.head()
```

```python
for index,row in df_data.iterrows():
    name = row['filename']
    label = row['label']
    filename = os.path.join('dataset',name)
    command = f"python inference_program.py -f {filename}"
    os.system(command)
    print('True Label: '+label)
```
