```python
import numpy as np
import os
from os import listdir
import librosa as lb
import pandas as pd
from collections import Counter
from imblearn.over_sampling import SMOTE
```

```python
files_loc = 'data_split/'
```

```python
img_train = []
label_train = []

to_hot_one = {'Bronchiectasis':0, 'Bronchiolitis':1, 'COPD':2, 'Healthy':3, 'Pneumonia':4, 'URTI':5}

for f in os.listdir(files_loc+'train/'):
    for i in os.listdir(files_loc+'train/'+f):
        filename = files_loc+'train/'+f+'/'+i
        print(f, i)
        sound, sample_rate = lb.load(filename)
        stft = lb.stft(y=sound, n_fft=512)
        chroma = np.mean(lb.feature.chroma_stft(S=stft, sr=sample_rate),axis=1)
        chroma_cqt = np.mean(lb.feature.chroma_cqt(y=sound, sr=sample_rate),axis=1)
        chroma_cens = np.mean(lb.feature.chroma_cens(y=sound, sr=sample_rate),axis=1)
        rms = np.mean(lb.feature.rms(y=sound),axis=1)
        mfccs = np.mean(lb.feature.mfcc(S=stft, sr=sample_rate),axis=1)
        mel = np.mean(lb.feature.melspectrogram(S=stft, sr=sample_rate),axis=1)
        centroid = np.mean(lb.feature.spectral_centroid(y=sound, sr=sample_rate, n_fft=512),axis=1)
        bandwidth = np.mean(lb.feature.spectral_bandwidth(y=sound, sr=sample_rate, n_fft=512),axis=1)
        contrast = np.mean(lb.feature.spectral_contrast(y=sound, sr=sample_rate, n_fft=512),axis=1)
        flatness = np.mean(lb.feature.spectral_flatness(y=sound, n_fft=512),axis=1)
        rolloff= np.mean(lb.feature.spectral_rolloff(y=sound, sr=sample_rate, n_fft=512),axis=1)
        poly= np.mean(lb.feature.poly_features(S=stft, sr=sample_rate),axis=1)
        tonnetz = np.mean(lb.feature.tonnetz(y=lb.effects.harmonic(sound), sr=sample_rate),axis=1)
        zcr= np.mean(lb.feature.zero_crossing_rate(y=sound),axis=1)
    
        concat = np.concatenate((chroma,chroma_cqt,chroma_cens,rms,mfccs,mel,centroid,bandwidth,contrast,flatness,rolloff,poly,tonnetz,zcr))
        img_train.append(concat)
        label_train.append(to_hot_one[f])
      

```
