```python
import numpy as np
import os
from os import listdir
import librosa as lb
import pandas as pd
from collections import Counter
from imblearn.over_sampling import SMOTE
```

```python
files_loc = 'data_split/'
```

```python
img_train = []
label_train = []

to_hot_one = {'Bronchiectasis':0, 'Bronchiolitis':1, 'COPD':2, 'Healthy':3, 'Pneumonia':4, 'URTI':5}

for f in os.listdir(files_loc+'train/'):
    for i in os.listdir(files_loc+'train/'+f):
      filename = files_loc+'train/'+f+'/'+i
      y,sr = lb.load(filename)
      mfcc = np.mean(lb.feature.mfcc(y=y,sr=sr),axis=1)
      img_train.append(mfcc)
      label_train.append(to_hot_one[f])
      print(f, i)
print('------DONE------')

```

```python
img_val = []
label_val = []

to_hot_one = {'Bronchiectasis':0, 'Bronchiolitis':1, 'COPD':2, 'Healthy':3, 'Pneumonia':4, 'URTI':5}

for f in os.listdir(files_loc+'val/'):
    for i in os.listdir(files_loc+'val/'+f):
      filename = files_loc+'val/'+f+'/'+i
      y,sr = lb.load(filename)
      mfcc = np.mean(lb.feature.mfcc(y=y,sr=sr),axis=1)
      img_val.append(mfcc)
      label_val.append(to_hot_one[f])
      print(f, i)

```

```python
img_test = []
label_test = []

to_hot_one = {'Bronchiectasis':0, 'Bronchiolitis':1, 'COPD':2, 'Healthy':3, 'Pneumonia':4, 'URTI':5}

for f in os.listdir(files_loc+'test/'):
    for i in os.listdir(files_loc+'test/'+f):
      filename = files_loc+'test/'+f+'/'+i
      y,sr = lb.load(filename)
      mfcc = np.mean(lb.feature.mfcc(y=y,sr=sr),axis=1)
      img_test.append(mfcc)
      label_test.append(to_hot_one[f])
      print(f, i)

```

```python
img_train = np.array(img_train)
label_train = np.array(label_train)

img_val = np.array(img_val)
label_val = np.array(label_val)

img_test = np.array(img_test)
label_test = np.array(label_test)
```

```python
print(img_train.shape)
print(img_val.shape)
print(img_test.shape)
```

```python
np.save("array/img_train.npy",img_train)
np.save("array/label_train.npy",label_train)

np.save("array/img_val.npy",img_val)
np.save("array/label_val.npy",label_val)

np.save("array/img_test.npy",img_test)
np.save("array/label_test.npy",label_test)
```
